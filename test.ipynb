```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
from random import *
from skrebate import ReliefF
# pip install skrebate
s = 'https://archive.ics.uci.edu/ml/'\
'machine-learning-databases/wine/wine.data'
#'machine-learning-databases/iris/iris.data'
print('From URL:', s)
df = pd.read_csv(s, header=None)
df.columns = ['Class label', 'Alcohol',
					'Malic acid', 'Ash',
					'Alcalinity of ash', 'Magnesium',
					'Total phenols', 'Flavanoids',
					'Nonflavanoid phenols',
					'Proanthocyanins',
					'Color intensity', 'Hue',
					'OD280/OD315 of diluted wines',
					'Proline']
print(df.info())
X = df.iloc[:,1:].values
L = df.iloc[:,0].values
r = ReliefF(n_neighbors=2, n_features_to_select=None)
M = r.fit(X, L)
for feature_name, feature_score in zip(df.drop('Class label', axis=1).columns,
                                       r.feature_importances_):
    print(feature_name, '\t', feature_score)
    
print(r.feature_importances_)
```
